```python

# 4- Identifying and handling the missing values

df.isnull().sum()
```

```python
df1 = df.copy()
```

```python
# FIRST MENTHOD: DROPNA

# summarize the shape of the raw data
print("Before:",df1.shape)

# drop rows with missing values
df1.dropna(inplace=True)

# summarize the shape of the data with missing rows removed
print("After:",df1.shape)
```

```python
# SOLUTION 2 : SECOND MENTHOD:FILLNA

df1 = df.copy()
```

```python

avg_fat = df1['Fat (g)'].mean()
df1['Fat (g)'] = df1['Fat (g)'].fillna(avg_fat)
print(df1.isnull().sum())
df1
```

```python
avg_cal = df1['Calories'].mean()
df1['Calories'] = df1['Calories'].fillna(avg_cal)
print(df1.isnull().sum())
df1
```

```python
avg_pro = df1['Protein (g)'].mean()
df1['Protein (g)'] = df1['Protein (g)'].fillna(avg_pro)
print(df1.isnull().sum())
df1
```

```python
avg_car = df1['Carbs (g)'].mean()
df1['Carbs (g)'] = df1['Carbs (g)'].fillna(avg_car)
print(df1.isnull().sum())
df1
```
